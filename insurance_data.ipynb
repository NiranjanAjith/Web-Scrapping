```python
import scrapy
from scrapy.crawler import CrawlerProcess
import pandas as pd
import os
```

```python
class CustomSpider(scrapy.Spider):
    name = 'custom_spider'
    start_urls = pd.read_csv('domains.csv')['Data'].tolist()

    def parse(self, response):
        items = []
        url_index = self.start_urls.index(response.url)
        
        # Select the main content of the page
        main_content = response.css('body').get()  # Adjust this selector to target main content more precisely

        # Exclude common ad elements
        ad_selectors = [
            '.ad',  # General ads
            '.ads',  # Multiple ads
            '.advertisement',  # Advertisement
            '.sponsored',  # Sponsored content
            '[id^="ad-"]',  # IDs starting with "ad-"
            '[class*="ad-"]',  # Classes containing "ad-"
            '[class*="ads"]',  # Classes containing "ads"
        ]

        # Remove ad elements
        for ad_selector in ad_selectors:
            ads = response.css(ad_selector)
            for ad in ads:
                ad.extract()

        # Collect the remaining content
        for paragraph in response.css('body p::text').getall():  # Adjust this selector as needed
            items.append(paragraph.strip())
        self.log(f"Items extracted: {items}")
        self.save_to_txt(items, url_index)

    def save_to_txt(self, items, index):
        # Define the directory and ensure it exists
        directory = 'insurance_data'
        if not os.path.exists(directory):
            os.makedirs(directory)

        # Define the filename with the index
        filename = os.path.join(directory, f'content_{index}.txt')

        # Write the items to the text file
        with open(filename, 'w') as file:
            for item in items:
                file.write(item + '\n')

        print(f"Content has been saved to {filename}")
```

```python
def run_spider():
    process = CrawlerProcess(settings={
       # 'LOG_LEVEL': 'DEBUG',# Add any custom settings here if needed
    })
    process.crawl(CustomSpider)
    process.start()
```

```python
# Run the Scrapy spider
run_spider()
```
